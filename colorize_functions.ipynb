```python
def hue_to_rgb(Hue,debug="False"):
    # input: Hue = int 0-359
    # output: H_rgb = array of 3 int
    
    # for the purposes of this function,
    # luminosity will always be 50%
        
    Hue=int(Hue)
    Lum=0.5
    Sat=1 #saturation; always use full saturation

    C= Sat*(1-abs(2*Lum-1))
    X= C*(1-abs((Hue/60)%2-1))
    m= Lum - (C/2)

    #Determine RGB
    rgb=[0,0,0]

    #R'
    if (120<=Hue<240):
        R_=0
    elif (60<=Hue<300):
        R_=X
    else:
        R_=C
    #G'
    if (60<=Hue<180):
        G_=C
    elif (Hue<240):
        G_=X
    else:
        G_=0
    #B'
    if (180<=Hue<300):
        B_=C
    elif (120<=Hue):
        B_=X
    else:
        B_=0

    rgb[0]=int((R_+m)*255)
    rgb[1]=int((G_+m)*255)
    rgb[2]=int((B_+m)*255)
    
    if debug==True:
        print("Hue,Lum,Sat,C,X,m: ",Hue,Lum,Sat,C,X,m)
        print("\tHue ",Hue," -> ",rgb)
    
    return rgb
```
